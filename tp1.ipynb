```python
image=cv2.imread("/content/drive/MyDrive/tp1/photo/_128806608_mediaitem128806607.jpg")
plt.imshow(image[...,::-1])
```

```python
gray_image= cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image,cmap="gray")
```

```python
image_blur=cv2.GaussianBlur(gray_image,(5,5),0.5)
plt.imshow(image_blur,cmap="gray")
```

le filtre bilatéral préserve les détails et réduit le bruit  mieux que le filtre gaussien

```python
image_filtree = cv2.bilateralFilter(image, 9, 75, 75)
plt.imshow(gray_image,cmap="gray")
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5, 5))
img_eroded = cv2.erode(gray_image, kernel)
plt.imshow(img_eroded,cmap="gray")
```

```python
img_dilated = cv2.dilate(gray_image, kernel)
plt.imshow(img_dilated,cmap="gray")
```

```python
img_closed = cv2.morphologyEx(gray_image, cv2.MORPH_CLOSE, kernel)
plt.imshow(img_closed,cmap="gray")
```

```python
img_opened = cv2.morphologyEx(gray_image, cv2.MORPH_OPEN, kernel)
plt.imshow(img_opened,cmap="gray")
```

les opérateurs morphologiques servent à améliorer la qualité et la clarté d'une image en fonction de l'objectif recherché.
